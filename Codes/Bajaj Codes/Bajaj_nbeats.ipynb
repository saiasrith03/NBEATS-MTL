```python
df = final_df.reset_index()
df
```

```python
df.drop('index',  axis = 1, inplace=True)
df
```

```python
df = df.set_index('Date')
df.index = pd.to_datetime(df.index)
```

```python
df.sort_index(inplace=True)
```

```python
df['Price'] = (df['Price'].str.split()).apply(lambda x: float(x[0].replace(',', '')))
df['Open'] = (df['Open'].str.split()).apply(lambda x: float(x[0].replace(',', '')))
df['High'] = (df['High'].str.split()).apply(lambda x: float(x[0].replace(',', '')))
df['Low'] = (df['Low'].str.split()).apply(lambda x: float(x[0].replace(',', '')))
df['Change %'] = (df['Change %'].str.split()).apply(lambda x: float(x[0].replace('%', '')))
df.info()
```

```python
df = df[['Price', 'Open', 'High', 'Low']].copy()
df
```

```python
df = df.dropna()
df
```

```python
df = df.resample('D').fillna(method='ffill')
```

```python
df
```

```python
# Ensure the 'Price' column is numeric
# test_df['Price'] = pd.to_numeric(test_df['Price'], errors='coerce')

# If there are any None values after conversion, this will remove those rows
# df.dropna(subset=['Price'], inplace=True)

window = 21
trd_days = 252
ann_factor = window / trd_days

# Calculate log returns
df['log_return'] = np.log(df['Price'] / df['Price'].shift(1))

# Calculate realized variance
rlz_var = df['log_return'].rolling(window=window).var() * ann_factor

# Calculate realized volatility
rlz_vol = np.sqrt(rlz_var)

# You might want to store the realized volatility in the DataFrame
df['realized_volatility'] = rlz_vol
```

```python
df = df.reset_index(drop=False)
df
```

```python
df = df.dropna()
df
```

```python
df['unique_id'] = '1'
df.rename(columns={'Date':'ds'}, inplace=True)
df.rename(columns={'realized_volatility':'y'}, inplace=True)
```
